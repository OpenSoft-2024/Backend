```python
from sentence_transformers import SentenceTransformer
import numpy as np
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
# model2=
```

```python
def cosine_similarity(a, b):
  return np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))

def get_embedding(text):
  embeddings = model.encode(text)
  return embeddings
```

```python
text1 = "Hello India, I am hjscbshsc"
text2 = "Hello, India"

print(cosine_similarity(get_embedding(text1), get_embedding(text2)))
```

```python
# Get embeddings for the first 100 rows for 'plot' and 'title' fields
embedding_collection = db['movie_Embedding']
embeddings_dict = {}

cursor = collection.find({}).limit(100)  # Limiting to first 100 rows
for document in cursor:
    movie_id = str(document.get('_id', ''))  # Assuming '_id' field contains the movie ID
    plot_text = document.get('fullplot', '')  # Assuming 'plot' field exists in the documents
    title_text = document.get('title', '')  # Assuming 'title' field exists in the documents
    plot_embedding = np.array(get_embedding(plot_text))  # Assuming get_embedding returns ndarray
    title_embedding = np.array(get_embedding(title_text))  # Assuming get_embedding returns ndarray
    embeddings_dict[movie_id] = {'plot': plot_embedding.tolist(), 'title': title_embedding.tolist()}


# Save the embeddings dictionary to the new collection
embedding_collection.insert_one(embeddings_dict)
```
