```python
from sentence_transformers import SentenceTransformer
import numpy as np
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
# model2=
```

```python
def cosine_similarity(a, b):
  return np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))

def get_embedding(text):
  embeddings = model.encode(text)
  return embeddings
```

```python
text1 = "Hello India, I am hjscbshsc"
text2 = "Hello, India"

print(cosine_similarity(get_embedding(text1), get_embedding(text2)))
```

```python
#plot semantic search
text = "misadventures buster left"

results = collection.aggregate([
  {"$vectorSearch": {
    "queryVector":np.array(get_embedding(text)).tolist(),
    "path": "plot_embedding",
    "numCandidates": 100,
    "limit": 10, #set it accordingly
    "index": "plot_embedding",
      }}
]);

for document in results:
    print(f'Movie Name: {document["title"]},\nMovie Plot: {document["plot"]}\n')

```

```python
#title semantic search
text = "idiot"

results = collection.aggregate([
  {"$vectorSearch": {
    "queryVector":np.array(get_embedding(text)).tolist(),
    "path": "title_embedding",
    "numCandidates": 100,
    "limit": 10, #set it accordingly
    "index": "vector_index",
      }}
]);

for document in results:
    print(f'Movie Name: {document["title"]},\nMovie Plot: {document["plot"]}\n')
```
